```python
import os
```

```python
num = input('Run on which core? (0 or 1) > ')

while True:
    while num not in ["0", "1"]:
        num = input('Invalid choice. Please enter 0 or 1')
    print(f'Running fib.py on core {num} with 30 terms')
    os.system(f"echo '30' | taskset -c {num} ./fib.py")
    num = input('Run on which core? (0 or 1) > ')
    
```
